# Hard Margin Linear SVM Classifier

```python
def hard_margin_linear_SVM(X, y):


    m,n = X.shape


    t = np.where(y == 0, -1, 1)


    model = gp.Model()

    w = model.addMVar(
        lb = float('-inf'),
        ub = float('inf'),
        vtype= GRB.CONTINUOUS,
        shape = n,
        name="w"
    )


    b = model.addVar(
        lb = float('-inf'),
        ub = float('inf'),
        vtype = GRB.CONTINUOUS,
        name = "b"
    )

    coeff = 0.5 * np.eye(n)
    model.setObjective(  w @ coeff @ w, GRB.MINIMIZE)
    
    
    model.addConstrs(t[i] * (w @ X[i, :] + b) >= 1 for i in range(m) )
    
    model.update()
    model.optimize()
    return w.X, b.X
```

```python
def predict(X, w, b):
    z=  X@w + b
    return np.where(z < 0, 0, 1)
```

```python
hard_margin_linear_SVM(X_train,y_train)
```

```python
def soft_margin_linear_SVM(X, y, C=0.5):


    m,n = X.shape


    t = np.where(y == 0, -1, 1)

    model = gp.Model()

    w = model.addMVar(
        lb = float('-inf'),
        ub = float('inf'),
        vtype= GRB.CONTINUOUS,
        shape = n,
        name="w"
    )
    
    z = model.addMVar(
        lb = 0, # enforce zeta >= 0 
        ub = float('inf'),
        vtype= GRB.CONTINUOUS,
        shape = m,
        name="z"
    )
    

    b = model.addVar(
        lb = float('-inf'),
        ub = float('inf'),
        vtype = GRB.CONTINUOUS,
        name = "b"
    )

    coeff_w = 0.5 * np.eye(n)
    coeff_z = C * np.ones(m)
    model.setObjective(  w @ coeff_w @ w + coeff_z @ z, GRB.MINIMIZE)
    
    model.addConstrs(z[i]+ t[i] * (w @ X[i, :] + b) >=1 for i in range(m) )
    
    model.update()
    model.optimize()
    return w.X, b.X, z.X
```

```python
params = np.linspace(0.001,1, 5)
accuracies = []

for c in params:
    w, b,z  = soft_margin_linear_SVM(X_train,y_train, C=c)
    print(z)
    y_pred = predict(X_val, w, b)
    accuracy_c = accuracy_score(y_val, y_pred)
    accuracies.append(accuracy_c)
    
```
