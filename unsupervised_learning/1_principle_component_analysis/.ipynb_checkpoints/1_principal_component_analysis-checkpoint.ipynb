```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler, scale
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("/home/josh/inde577/schaffer_inde_577/data/rb_tx_data.csv")
df = df.set_index("TransactionID")

df = df.replace(np.nan,0)

df = df.loc[(df['TotalCost']>0) & 
            (df['RepairCost']>0) &
            (df['TurnaroundTime']>0) &
            (df['TurnaroundTime']<38) 
            ]

X = df.loc[:, df.columns[14:] ].values
#X = StandardScaler().fit_transform(X)
X = scale(X)
y = np.where(df['DayCreated'] == df['DayCompleted'], 0, 1)
y = np.where(df['TransactionType'] == 'outpatient', 0, 1)

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state = 42)

X_train.shape
```

```python
U, sigma, V_T = np.linalg.svd(X)
```

```python
X_train.shape
```

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(X)

print(f"pca.explained_variance_ratio_ = {pca.explained_variance_ratio_}")
```

```python
PCs = range(X_train.shape[1])
plt.figure(figsize=(10,8))
plt.plot(PCs, pca.explained_variance_ratio_)
```

```python
PCs = range(X_train.shape[1])
cumulative_variance = np.cumsum( pca.explained_variance_ratio_)
plt.figure(figsize=(10,8))
plt.plot(PCs, cumulative_variance)
```

```python
S = ( 1/X.shape[0] ) * X @ X.T
S.shape
```

```python
# the entires of sigma are the square roots of the eigenvalues of S 
sigma
```

```python
# Variance described by each prinicpal component is given by the quotient of that respective eigvenalue and the trace of S
# note that the trace of S is equivalent to the sum of the eigenvalues
pc_variance = sigma **2 / (sigma**2).sum()
pc_variance
```

```python
cumulative_pc_variance = np.cumsum(pc_variance)
```

```python
PCs = range(X_train.shape[1])
plt.figure(figsize=(10,8))
plt.plot(PCs, cumulative_pc_variance)
```
